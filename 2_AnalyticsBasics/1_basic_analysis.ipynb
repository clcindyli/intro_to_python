We've done some simple aggregate functions e.g. max, average, median, etc. but now we can start looking at relationships between variables.

```python
#import pandas
import pandas as pd
```

```python
#read in data
df = pd.read_csv("GSS.2006.csv.xls")
df.head()
```

### Cross Tabs

We use cross tabulations to see the distribution for different variables.

```python
#we can also use the crosstab function to see percentages for the variables we want to see

tabulation = pd.crosstab(index=df["jobsat"],  # Make a crosstab
                              columns="count")     
def compute_percentage(x):
      pct = float(x / tabulation['count'].sum()) * 100
      return round(pct, 2)

tabulation['percentage'] = tabulation.apply(compute_percentage, axis=1)

tabulation
```

We can also look at the cross-section between two variables

```python
res = pd.crosstab(df.degree, df.famgen)
res.astype('float').div(res.sum(axis=0), axis=1)
```
