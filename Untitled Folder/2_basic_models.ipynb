### Recoding Data

Sometimes our data is not in a format that is easy to work with and we need to recode it

```python
# map indicator variables - we map them to make the numbers strings
df['degree'].replace({0: 'Lt High School', 1: 'High School', 2:'Junior College', 3: 'Bachelor', 4:'Graduate'}\
                         , inplace=True)

# we can create a crosstab to see our updated data & get percentages

#define comp_pct function
def comp_pct(x):
      pct = float(x / tab['count'].sum()) * 100
      return round(pct, 2)


tab = pd.crosstab(index=df["degree"],  # Make a crosstab
                              columns="count")     

tab['percentage'] = tab.apply(comp_pct, axis=1)

tab
```

```python
# we can also put a continuous variable into bins, using different cutoffs

def binning(col, cut_points, labels=None):
  #Define min and max values:
  minval = col.min()
  maxval = col.max()

  #create list by adding min and max to cut_points
  break_points = [minval] + cut_points + [maxval]

  #if no labels provided, use default labels 0 ... (n-1)
  if not labels:
    labels = range(len(cut_points)+1)

  #Binning using cut function of pandas
  colBin = pd.cut(col,bins=break_points,labels=labels,include_lowest=True)
  return colBin

#Binning attend:

cut_points = [1, 2, 4, 7]
labels = ["small_family","medium_family","van_family","large_family", "xl_family"]
df["sibs_cut"] = binning(df["sibs"], cut_points, labels)
```

```python
#look at our new bins

tab_sibs = pd.crosstab(index=df["sibs_cut"], columns="count")     
tab_sibs['percentage'] = tab_sibs.apply(comp_pct, axis=1)

tab_sibs
```

### Linear Regression

Simple linear regressions to find relationships. More complex models require more thought - but you can easily apply other types of models within the statsmodel package in a similar fashion.

```python
#define your variables
x = df.population
y = df.total_households
```

```python
#give your model a constant
x = sm.add_constant(x)
```

```python
#OLS Model
model = sm.OLS(y, x)
results = model.fit()
results.summary()
```
